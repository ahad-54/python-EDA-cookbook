# **Histogram:**
A histogram in Seaborn is a bar-like plot that shows how often different ranges of values occur in a dataset.

# **Code Logic :-**
The below code generates a histogram plot which shows how many items are in the dataset with particular calorie count.

```python
plt.figure(figsize= (12,6))

ax = sns.histplot( data = md1, x= md1["Calories"])
ax.set_xlabel('Calorie Count',fontsize = 15, fontweight='bold')
ax.set_ylabel('No. of items', fontsize = 15, fontweight='bold')
ax.set_title('No. of items in specific calorie bracket', fontsize= 20)
```

# **Box Plot :**
A boxplot finds and shows the summary of a dataset’s spread (minimum, maximum, median, quartiles, and outliers) in one compact graph.

# **Code Logic :-**
This code draws a boxplot of the Carbs column from my dataset.

The blue box shows the middle 50% of Carbs values and the line inside is the median Carb and the whiskers show the minimum and maximum range, and any dots outside are outliers.

This graph helps us to find the average Carb level, the spread of Carb across items, and if some foods have unusually high or low Carbs compared to others.

```python
plt.figure(figsize= (10,6))

ax = sns.boxplot(data = md1, x= md1["Carbs "])
ax.set_xlabel('Carbs(g)',fontsize = 15, fontweight='bold')
ax.set_title('Distribution of Carbs in McDonalds items', fontsize= 20)
plt.ticklabel_format(style='plain', axis='x')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# **VioLinPlot :**
It is just similar to boxplot but we can say it is more fansy and easy to visualize.

A violin plot combines a boxplot with a density curve, so it not only shows the median and quartiles (like a boxplot) but also the shape of the data distribution (where values are more frequent).

# **Code Logic :-**
Below I created a violin plot of the Sugars column from my dataset.

According to my dataset and from this graph, we can find the average Sugar levels, the spread of Sugars across items, and whether the data is skewed or has multiple peaks.

```python
plt.figure(figsize= (10,6))

ax = sns.violinplot(data = md1, x= md1["Sugars "])
ax.set_xlabel('Sugars(g)',fontsize = 15, fontweight='bold')
ax.set_title('Distribuion of sugar levels in McDonalds products', fontsize= 20)
plt.ticklabel_format(style='plain', axis='x')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# **Creating a bar plot using countplot method in Seaborn :**


```python

plt.figure(figsize=(10,5))
ax = sns.countplot(x='Type', data=md1)
ax.set_title('No. of Types of McDonalds Items', fontsize=16, fontweight='bold')
ax.set_xlabel('Type of items', fontsize=12, fontweight='bold')
ax.set_ylabel('Count', fontsize=12, fontweight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **Pie-chart in matplotlib**

The pie chart helps us quickly see which the data  values most commonlt and easily , also how the items are distributed across the dataset levels.

# **Code Logic:**
The below code distributes the type of items along a pie-chart using a percentage of the no.of different type of items spread accross my dataset.

```python
plt.pie(group['Count'], labels=group['Type'], autopct='%1.0f%%',
        startangle=140, colors=sns.color_palette("pastel"),
        wedgeprops={'edgecolor': 'black'})
plt.title('Type of McDonalds Items', fontweight='bold')
plt.show()
```
